```python
from tensorflow.keras import models, layers
from tensorflow.keras.utils import plot_model

# Define the LeNet-5 model
def lenet5_model():
    model = models.Sequential()
    model.add(layers.Conv2D(6, (5, 5), activation='relu', input_shape=(32, 32, 1)))
    model.add(layers.AveragePooling2D())

    model.add(layers.Conv2D(16, (5, 5), activation='relu'))
    model.add(layers.AveragePooling2D())

    model.add(layers.Flatten())
    model.add(layers.Dense(120, activation='relu'))
    model.add(layers.Dense(84, activation='relu'))
    model.add(layers.Dense(10, activation='softmax'))  # 10 classes

    return model

# Create the LeNet-5 model
lenet5 = lenet5_model()

# Model summary
lenet5.summary()
```

```python
# Plot the model
plot_model(lenet5, to_file='lenet5_model.png', show_shapes=True, show_layer_names=True)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 定义原始的LeNet-5
def original_lenet5(input_shape):
    model = Sequential()
    model.add(Conv2D(6, (5, 5), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(16, (5, 5), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(120, activation='relu'))
    model.add(Dense(84, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

# 定义一个简化版的LeNet-5，减少卷积核的数量
def simplified_lenet5(input_shape):
    model = Sequential()
    model.add(Conv2D(4, (5, 5), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(8, (5, 5), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(120, activation='relu'))
    model.add(Dense(84, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    return model

# 定义一个带有Dropout层的LeNet-5版本
def lenet5_with_dropout(input_shape, dropout_rate=0.5):
    model = Sequential()
    model.add(Conv2D(6, (5, 5), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(16, (5, 5), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(120, activation='relu'))
    model.add(Dense(84, activation='relu'))
    model.add(tf.keras.layers.Dropout(dropout_rate))
    model.add(Dense(10, activation='softmax'))
    return model

# 输入图像的形状
input_shape = (28, 28, 1)  # 这里假设输入图像是32x32像素的RGB图像

# 创建模型并查看其summary
original_model = original_lenet5(input_shape)
simplified_model = simplified_lenet5(input_shape)
dropout_model = lenet5_with_dropout(input_shape)

original_model.summary()
simplified_model.summary()
dropout_model.summary()
```
