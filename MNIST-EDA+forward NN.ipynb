# DL model in Keras

Sequential is useful for stacking layers where each layer has one input tensor and one output tensor. Layers are functions with a known mathematical structure that can be reused and have trainable variables. Most TensorFlow models are composed of layers. This model uses the Flatten, Dense, and Dropout layers.

For each example, the model returns a vector of logits or log-odds scores, one for each class.

```python
# Import the required TensorFlow and Keras libraries
import tensorflow as tf

# Create a Sequential model, which represents a linear stack of layers.
model = tf.keras.models.Sequential([
    # Flatten layer: Reshapes the input data from 28x28 to a 1D array of size 784.
    tf.keras.layers.Flatten(input_shape=(28, 28)),

    # Dense layer: Fully connected layer with 128 neurons and ReLU activation function.
    tf.keras.layers.Dense(128, activation='relu'),

    # Dropout layer: Applies dropout regularization with a rate of 0.2.
    tf.keras.layers.Dropout(0.2),

    # Dense layer: Fully connected layer with 10 neurons (output layer).
    # Note: There's no activation function specified, which implies a linear activation.
    tf.keras.layers.Dense(10)
])
```

Explanation:

- import tensorflow as tf: Import the TensorFlow library, typically aliased as tf.

- tf.keras.models.Sequential([]): Create a Sequential model. This type of model allows you to define a neural network as a linear stack of layers, where data flows sequentially through each layer.

- tf.keras.layers.Flatten(input_shape=(28, 28)): The first layer is a Flatten layer. It takes the input data, which is assumed to be 28x28 pixels (common for image data), and flattens it into a 1D array of size 784 (28x28 = 784). This is necessary because many neural networks expect a flat input, especially for fully connected layers.

- tf.keras.layers.Dense(128, activation='relu'): The second layer is a Dense (fully connected) layer with 128 neurons. It uses the Rectified Linear Unit (ReLU) activation function, which introduces non-linearity to the model. ReLU is commonly used in hidden layers to capture complex patterns in the data.

- tf.keras.layers.Dropout(0.2): The Dropout layer helps prevent overfitting by randomly setting a fraction (in this case, 20%) of input units to 0 during each training step. This reduces the risk of the model relying too heavily on any individual neuron.

- tf.keras.layers.Dense(10): The final layer is another Dense layer with 10 neurons. This is the output layer of the network, and there's no activation function specified. In this case, it's treated as a linear layer, and the raw scores from this layer will be used in a later step to compute probabilities using a softmax activation for classification tasks.


This code defines a simple feedforward neural network with one hidden layer, suitable for tasks like image classification. You can further compile the model, specify a loss function, an optimizer, and train it on your dataset.

# Train the model in PyTorch with the tf dataset

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np

# Define a PyTorch model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x
```

- class Net(nn.Module): Defines a new class Net that inherits from nn.Module. This class represents the neural network model.

- def __init__(self):: The constructor method for the Net class.

- super(Net, self).__init__(): Calls the constructor of the parent class (nn.Module) to initialize the Net class.

- self.flatten = nn.Flatten(): Creates a Flatten layer that will be used to flatten the input data into a 1D vector. This is often necessary when processing image data.

- self.fc1 = nn.Linear(28 * 28, 128): Defines the first fully connected (linear) layer with 28*28 input features and 128 output features. This is followed by:

- self.relu = nn.ReLU(): A Rectified Linear Unit (ReLU) activation function layer to introduce non-linearity.
self.dropout = nn.Dropout(0.2): A dropout layer with a dropout rate of 20%. Dropout is used for regularization to prevent overfitting.
- self.fc2 = nn.Linear(128, 10): Defines the second fully connected layer with 128 input features and 10 output features. This corresponds to the number of classes in the classification task.

The forward method defines the forward pass of the model. It specifies how input data x should be processed as it passes through the layers defined in the constructor.
The input is flattened using self.flatten(x), then passed through self.fc1, followed by the ReLU activation, dropout, and the final fully connected layer self.fc2.
The output of the forward pass is returned, which represents the model's predictions.

This code defines a simple feedforward neural network for a classification task. The model consists of an input layer (flattening layer), a hidden layer (with ReLU activation and dropout), and an output layer. Once the model is defined, it can be trained on your dataset using an appropriate loss function and optimizer. Additionally, it visualizes the training process using Matplotlib.

```python
# Initialize the model
model = Net()

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5
train_losses = []

for epoch in range(num_epochs):
    running_loss = 0.0
    for i in range(len(x_train)):
        inputs = torch.Tensor(x_train[i]).unsqueeze(0)
        labels = torch.LongTensor([y_train[i]])

        optimizer.zero_grad()
        outputs = model(inputs)

        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Calculate and store the average loss for this epoch
    train_loss = running_loss / len(x_train)
    train_losses.append(train_loss)

    print(f"Epoch [{epoch + 1}/{num_epochs}] - Loss: {train_loss:.4f}")

# Visualize the training process
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.show()
```

The provided code segment is for training a neural network model using PyTorch and visualizing the training process using Matplotlib. Let's break down the code step by step:

- model = Net(): In this line, an instance of the Net class (which represents the neural network model) is created. This is where the model architecture is defined.

- criterion = nn.CrossEntropyLoss(): Here, the loss function is defined. nn.CrossEntropyLoss() is commonly used for classification tasks. It computes the cross-entropy loss between predicted class scores and true labels.

- optimizer = optim.Adam(model.parameters(), lr=0.001): An optimizer is defined to update the model's weights during training. In this case, the Adam optimizer is used with a learning rate of 0.001. The model.parameters() function provides all the model's trainable parameters (weights and biases) to the optimizer.

- num_epochs = 5: This variable specifies the number of training epochs, which determines how many times the entire training dataset will be used for training.

- train_losses = []: An empty list is created to store the training loss values for each epoch. This will be used for visualization.

Training Loop:

- for epoch in range(num_epochs):: This loop iterates over the specified number of epochs.

- running_loss = 0.0: running_loss is initialized to zero. It will be used to accumulate the loss during each epoch.

- for i in range(len(x_train)):: This inner loop iterates over the training data. It goes through each training example in the dataset.

- inputs = torch.Tensor(x_train[i]).unsqueeze(0): For each training example, the input data is converted to a PyTorch tensor, and unsqueeze(0) is used to add an extra dimension at the beginning. This is necessary because PyTorch expects batched data even if the batch size is 1.

- labels = torch.LongTensor([y_train[i]]): Similarly, the target labels are converted to a PyTorch tensor.

- optimizer.zero_grad(): Before computing gradients, the gradients from the previous step are cleared. This is necessary because PyTorch accumulates gradients by default.

- outputs = model(inputs): The input data is passed through the neural network model to obtain predictions (outputs).

- loss = criterion(outputs, labels): The cross-entropy loss is calculated by comparing the model's predictions (outputs) with the ground truth labels (labels).

- loss.backward(): The gradients of the model's parameters with respect to the loss are computed using backpropagation.

- optimizer.step(): The optimizer updates the model's parameters based on the computed gradients.

- running_loss += loss.item(): The current batch's loss is added to the running loss.

At the end of each epoch, the average loss for that epoch is calculated and stored in the train_losses list.

- print(f"Epoch [{epoch + 1}/{num_epochs}] - Loss: {train_loss:.4f}"): The training loss for the current epoch is printed to the console.

After training is complete, the training loss values for each epoch are plotted using Matplotlib to visualize the training process.

This code demonstrates the training loop for a neural network, which involves forward and backward passes, gradient updates, and the accumulation of loss values over epochs. The final visualization shows how the training loss changes over the course of training, helping to assess the model's convergence.

# Train the model in PyTorch with PyTorch MNIST dataset

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torchvision.datasets import MNIST
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Define a custom neural network model in PyTorch
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28 * 28, 128)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Load MNIST dataset and normalize pixel values
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_dataset = MNIST(root='./data', train=True, transform=transform, download=True)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# Create an instance of the model
model = Net()

# Define loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 5
losses = []

for epoch in range(epochs):
    running_loss = 0.0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    losses.append(running_loss)

# Visualize the training process
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('Training Loss Over Epochs')
plt.show()
```
